### Insertion 
    Time Complexity:O(log n)
    Space Complexity:O(log n)(beacuse of recursion which uses stack)

```python
class Heap:
    def __init__(self, size):
        self.cutomList=(size+1)*[None]
        # make a list of length size +1 and fill it completely with 0  or None
        self.heapSize=0
        self.maxSize=size+1
        
def peek(root):
    if root  is None:
        return
    else:
        return root.cutomList[1]
    
def sizeofHeap(root):
       if not root:
            return
       return root.heapSize

def LevelOrder(root):
    if not root:
        return
    else:
        for i in range (1, root.heapSize+1):
            print (root.cutomList[i])
            
            
def heapifyTreeInsert(root, index, heaptype):
    parentIndex=int(index/2)
    if index <=1:
        return
    
    if heaptype =="Min":
        if root.cutomList[index]<root.cutomList[parentIndex]:
            temp=root.cutomList[index]
            root.cutomList[index]=root.cutomList[parentIndex]
            root.cutomList[parentIndex]=temp
    heapifyTreeInsert(root,parentIndex,heaptype)
    
    if heaptype =="Max":
        if root.cutomList[index]>root.cutomList[parentIndex]:
            temp=root.cutomList[index]
            root.cutomList[index]=root.cutomList[parentIndex]
            root.cutomList[parentIndex]=temp
    heapifyTreeInsert(root,parentIndex,heaptype)
            
def InsertNode(root, nodeValue, heaptype):
    if root.heapSize+1==root.maxSize:
        return ("Binary heap is full")
    else:
        root.cutomList[root.heapSize+1]=nodeValue
        root.heapSize+=1
        heapifyTreeInsert(root, root.heapSize, heaptype)
        return ("Value inserted")
        
newBinaryHeap=Heap(5)

InsertNode(newBinaryHeap,5 , "Min")
InsertNode(newBinaryHeap, 4, "Min")
InsertNode(newBinaryHeap, 2, "Min")
InsertNode(newBinaryHeap, 7, "Min")
InsertNode(newBinaryHeap, 4.5, "Min")
LevelOrder(newBinaryHeap)
InsertNode(newBinaryHeap, 78, "Min")
```

### Extract node
    Time Complexity: O(log n)
    Space Complexity:O(log n)

```python
class Heap:
    def __init__(self, size):
        self.cutomList=(size+1)*[None]
        # make a list of length size +1 and fill it completely with 0  or None
        self.heapSize=0
        self.maxSize=size+1
        
def peek(root):
    if root  is None:
        return
    else:
        return root.cutomList[1]
    
def sizeofHeap(root):
       if not root:
            return
       return root.heapSize

def LevelOrder(root):
    if root is None:
        return
    else:
        for i in range (1, root.heapSize+1):
            print (root.cutomList[i])
            
            
def heapifyTreeInsert(root, index, heaptype):
    parentIndex=int(index/2)
    if index <=1:
        return
    
    if heaptype =="Min":
        if root.cutomList[index]<root.cutomList[parentIndex]:
            temp=root.cutomList[index]
            root.cutomList[index]=root.cutomList[parentIndex]
            root.cutomList[parentIndex]=temp
    heapifyTreeInsert(root,parentIndex,heaptype)
    
    if heaptype =="Max":
        if root.cutomList[index]>root.cutomList[parentIndex]:
            temp=root.cutomList[index]
            root.cutomList[index]=root.cutomList[parentIndex]
            root.cutomList[parentIndex]=temp
    heapifyTreeInsert(root,parentIndex,heaptype)
            
def InsertNode(root, nodeValue, heaptype):
    if root.heapSize+1==root.maxSize:
        return ("Binary heap is full")
    else:
        root.cutomList[root.heapSize+1]=nodeValue
        root.heapSize+=1
        heapifyTreeInsert(root, root.heapSize, heaptype)
        return ("Value inserted")
        
def extract(root, index, heaptype):
    leftIndex=index*2
    rightIndex=index*2+1
    swapChild=0
    if root.heapSize<leftIndex:
        return 
    elif root.heapSize==leftIndex:
        if heaptype=="Min":
            if root.cutomList[index]>root.cutomList[leftIndex]:
                temp=root.cutomList[index]
                root.cutomList[index]=root.cutomList[leftIndex]
                root.cutomList[leftIndex]=temp
            return 
        else:
            if root.cutomList[index]<root.cutomList[leftIndex]:
                temp=root.cutomList[index]
                root.cutomList[index]=root.cutomList[leftIndex]
                root.cutomList[leftIndex]=temp
            return 
    
    else:
        if heaptype == "Min":
            if root.cutomList[leftIndex]< root.cutomList[rightIndex]:
                swapChild=leftIndex
            else:
                swapChild=rightIndex
            
            if root.cutomList[index]>root.cutomList[swapChild]:
                temp=root.cutomList[index]
                root.cutomList[index]=root.cutomList[swapChild]
                root.cutomList[swapChild]=temp
                
        else:
            if root.cutomList[leftIndex]> root.cutomList[rightIndex]:
                swapChild=leftIndex
            else:
                swapChild=rightIndex
            
            if root.cutomList[index] < root.cutomList[swapChild]:
                temp=root.cutomList[index]
                root.cutomList[index]=root.cutomList[swapChild]
                root.cutomList[swapChild]=temp
                
        extract(root, swapChild,heaptype)
            
            
def extract_node(root, heaptype):
    if root.heapSize==0:
        return 
    else:
        extractedNode=root.cutomList[1]
        root.cutomList[1]=root.cutomList[root.heapSize]
        root.cutomList[root.heapSize]=None
        root.heapSize-=1
        extract(root,1,heaptype)
        return extractedNode        
        
def del_heap(root):
    root.cutomList=None
    print("Deleted")
        
newBinaryHeap=Heap(5)
InsertNode(newBinaryHeap,5 , "Min")
InsertNode(newBinaryHeap, 4, "Min")
InsertNode(newBinaryHeap, 2, "Min")
InsertNode(newBinaryHeap, 7, "Min")
InsertNode(newBinaryHeap, 4.5, "Min")
LevelOrder(newBinaryHeap)
InsertNode(newBinaryHeap, 78, "Min")

print(extract_node(newBinaryHeap,"Min"))
print()
LevelOrder(newBinaryHeap)
print(extract_node(newBinaryHeap,"Min"))
print(extract_node(newBinaryHeap,"Min"))
print(extract_node(newBinaryHeap,"Min"))

del_heap(newBinaryHeap)
#LevelOrder(newBinaryHeap)
```
